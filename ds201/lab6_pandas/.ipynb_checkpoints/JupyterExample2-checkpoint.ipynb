```python
import pandas as pd
import matplotlib.pyplot as plt

dogs = pd.read_csv("dogs.csv")
dogs
```

```python
dogs["Height"].hist()
plt.show()
```

```python
avg_weight_breed = dogs.groupby("Breed")["Weight"].mean()
avg_weight_breed
```

```python
avg_weight_breed.plot(kind="bar", title="Mean Weight By Dog Breed")
plt.show()
```

```python
dogs.plot(x="Weight", y="Height", kind="scatter")
plt.show()
```

```python
dogs[dogs["Color"] == "Brown"]["Height"].hist(alpha=0.7)
dogs[dogs["Color"] == "Black"]["Height"].hist(alpha=0.7)
plt.legend(["Brown", "Black"])
plt.show()
```

```python
dogs["BMI"] = dogs["Weight"] / (dogs["Height"] / 100) ** 2
dogs
```
